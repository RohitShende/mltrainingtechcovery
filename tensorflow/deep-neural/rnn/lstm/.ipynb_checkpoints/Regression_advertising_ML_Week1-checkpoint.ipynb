```python
# Python program to run Ordinary Least-Squares Linear Regression on Advertising dataset
# Label (or dependent variable) = Sales
# Explanatory Variables = TV, Radio, and _Newspaper

#importing all the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
```

```python
#load data into dataframe for data cleaning

df = pd.read_csv("Advertising.csv")
df.head(5)

# check if there is any null value or any NaN values

#print (df.isnull().sum()) 
X = df[['TV', 'Radio', 'Newspaper']]
y = df['Sales']
```

```python
#Splitting train/test data in ratio of 75-25
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state = 0)

#Preprocessing data for regression
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)

#visualising training data
from matplotlib import cm
from pandas.plotting import scatter_matrix
from pandas import concat

%matplotlib inline
cmap = cm.get_cmap('gnuplot')
#converting X_train_scaled to dataframe for the below function to work
X_train_scaled1 = pd.DataFrame(X_train_scaled, columns=['TV', 'Radio', 'Newspaper'])
df_scattermatrix = concat([X_train_scaled1,y], axis = 1)
pd.plotting.scatter_matrix(df_scattermatrix, alpha = 0.5,figsize = (20,15), marker = 'o', diagonal = 'kde', cmap = cmap)
# create a scatter matrix from the dataframe, color by y_train
# example from stackoverflow - https://stackoverflow.com/questions/55394041/module-pandas-has-no-attribute-scatter-matrix
# pandas function documentation - https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.plotting.scatter_matrix.html
```

```python
#LR classifier
reg_clf = LinearRegression()
reg_clf.fit(X_train_scaled, y_train)
r2_score_test = reg_clf.score(X_test_scaled, y_test)
r2_score_train = reg_clf.score(X_train_scaled, y_train)
```

```python
print("Linear model coefficients are:",reg_clf.coef_)
print('\n')
print("linear model intercept is:",reg_clf.intercept_)
print('\n')
print("how close the data are to the fitted regression line or what is the coefficient of determination...")
print("R-squared score (testing)", r2_score_test)
print("R-squared score (training)", r2_score_train)
```
